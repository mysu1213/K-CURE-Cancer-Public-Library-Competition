# 1. 데이터 불러오기

```python
df_cc_rgst = pd.read_csv('./PP_csv/cs_cc_rgst.csv') # 21773 rows × 36 columns
df_cs_cc_rec = pd.read_csv('./PP_csv/cs_cc_rec.csv')
df_cs_cc_nrec = pd.read_csv('./PP_csv/cs_cc_nrec.csv')
df_cc_ckp = pd.read_csv('./PP_csv/cs_cc_g1e_fin.csv')
```

```python
df_cc_ckp.isnull().sum()
```

```python
df_cc_rgst_id = df_cc_rgst[['SN_KEY', 'FDX']] # 초진연월 붙이기
df_cc_ckp = pd.merge(df_cc_rgst_id, df_cc_ckp, how = 'inner', on = ['SN_KEY']) # 초진연월 추가
df_cc_ckp
```

```python
df_cc_ckp.isnull().sum()
```

---

##### Label 1 - Table

```python
## 재발피험자 불러오기 ###
cust_id = list(df_cs_cc_rec['SN_KEY'])
df_cc_ckp_rec = df_cc_ckp[df_cc_ckp['SN_KEY'].isin(cust_id)]
df_cc_ckp_rec
```

```python
## 재발 테이블 합치기###
tmp = pd.merge(df_cs_cc_rec, df_cc_ckp_rec, how = 'inner', on=['SN_KEY'])
tmp
```

```python
tmp['col_1_rec'] = (tmp['TIME_OF_REC_START'].str.slice(0, 4)).astype('int32') # 근접
tmp['col1_result'] = abs(tmp['col_1_rec'] - tmp['EXMD_BZ_YYYY'])
tmp_result = tmp[tmp.groupby('SN_KEY')['col1_result'].transform(min) == tmp['col1_result']]
tmp_result = tmp_result.drop_duplicates(subset='SN_KEY')
tmp_result
```

```python
df_cs_cc_rec = pd.merge(df_cs_cc_rec, tmp_result, how = 'outer', on = ['SN_KEY', 'LABEL', 'RES', 'SOPR_Y_CNT', 'TIME_OF_NREC_END', 'TIME_OF_REC_START', 'TIME_OF_REC'])
df_cs_cc_rec = df_cs_cc_rec.drop(['col_1_rec', 'col1_result'], axis = 'columns')
df_cs_cc_rec
```

```python
df_cs_cc_rec.to_csv('./PP_csv/cs_cc_rec_2.csv')
```

##### Label 0

```python
## 재발x 피험자 불러오기 ###
cust_id = list(df_cs_cc_nrec['SN_KEY'])
df_cc_ckp_nrec = df_cc_ckp[df_cc_ckp['SN_KEY'].isin(cust_id)]
# df_cc_ckp_nrec.head(40)
```

```python
df_cc_ckp_nrec['FDX_2'] = df_cc_ckp_nrec['FDX'].astype('str')
df_cc_ckp_nrec['FDX_2'] = df_cc_ckp_nrec['FDX_2'].str.slice(0, 4).astype('int32')
```

```python
df_cc_ckp_nrec['col_result'] = abs(df_cc_ckp_nrec['FDX_2'] - df_cc_ckp_nrec['EXMD_BZ_YYYY'])
```

```python
df_cc_ckp_nrec[df_cc_ckp_nrec.groupby('SN_KEY')['col_result'].transform(min) == df_cc_ckp_nrec['col_result']]
df_cc_ckp_nrec = df_cc_ckp_nrec.drop_duplicates(subset='SN_KEY')
```

```python
df_cs_cc_nrec = pd.merge(df_cs_cc_nrec, df_cc_ckp_nrec, how = 'outer', on = ['SN_KEY'])
df_cs_cc_nrec
```

```python
df_cs_cc_nrec = df_cs_cc_nrec.drop(['FDX_2', 'col_result'],axis= 'columns')
df_cs_cc_nrec
```

```python
df_cs_cc_nrec.to_csv('./PP_csv/cs_cc_nrec_2.csv')
```
